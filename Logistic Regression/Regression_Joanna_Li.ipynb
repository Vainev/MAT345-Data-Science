## 1) Linear regression

```python
# import all needed libraries
import pandas as pd
import numpy as np
import math
```

```python
trainingSet = pd.read_excel('Project3.xlsx', sheet_name='Training')
```

```python
x = np.c_[trainingSet.iloc[:, :3]]
y = np.c_[trainingSet.loc[:, 'Course Grade']]
```

```python
# augment x 
xAug = np.concatenate(((np.ones(x.shape[0]))[:, np.newaxis], x), axis=1)
```

```python
# find x ^ T * X
xTranX = np.matmul(xAug.T, xAug)

# find x^T * y
xTranY = np.matmul(xAug.T, y)

# find (x^T * x) ^-1
xTranXInverse =  np.linalg.inv(xTranX)
```

```python
weight = np.matmul(xTranXInverse, xTranY)
```

### 3) Run Logistic Regression Algorithm with Gradient Descent

```python
class LogisticRegression:
    def __init__(self, x, y, learningRate, iterations):
        self.x = x
        self.y = y
        self.w = None
        self.lr = learningRate
        self.steps = iterations
        
    # initalize weights to zero
    def initialize_weights(self, dim):
        self.w = np.zeros((len(dim)), dtype = float)
        
    def sigmoid(self, z):
        sig = 1/(1+np.exp(-z))
        return sig
    
    # gradient calculation
    def gradient(self, x, y):
        z = np.dot(self.w, x)
        grad = -x * y*(1- self.sigmoid(z))
        return grad
    
    # logistic regression with gradient descent
    def fit(self):        
        for i in range(self.steps):
            grad = 0
            for idx in range(len(self.x)):
                grad += self.gradient(self.x[idx], self.y[idx])

            # update weight
            self.w = self.w - self.lr * grad/len(self.x)

    def probability(self, X, w):
        z =  np.dot(w, X.T)
        predict = []
        for i in range(len(z)):
            sig = 0
            if z[i] <= 70:
                sig = self.sigmoid(-z[i])
            else:
                sig = self.sigmoid(z[i])
                
            if sig < 0.5:
                predict.append(0)
            else:
                predict.append(sig)
                
        return pd.DataFrame(predict, columns={'Probability of Passing'})
```

```python
regressor = LogisticRegression(xTrain, yTrain, 0.01, 5000)
regressor.initialize_weights(xTrain[0])
```

```python
regressor.fit()
```

### Final Weight

```python
print(regressor.w)
```
